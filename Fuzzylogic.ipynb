```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 
time = ctrl.Antecedent(np.arange(10,181,1),'time')
rice = ctrl.Antecedent(np.arange(100,1501,1),'rice')
pow = ctrl.Consequent(np.arange(10,101,1),'pow')
time['vfast'] = fuzz.trimf(time.universe,[10,10,20])
time['fast'] = fuzz.trimf(time.universe,[10,20,30])
time['medium'] = fuzz.trimf(time.universe,[20,30,40])
time['slow'] = fuzz.trimf(time.universe,[30,40,50])
time['vslow'] = fuzz.trapmf(time.universe,[40,50,180,180])
rice['vless'] = fuzz.trimf(rice.universe,[100,100,250])
rice['less'] = fuzz.trimf(rice.universe,[100,250,550])
rice['avg'] = fuzz.trimf(rice.universe,[250,550,800])
rice['much'] = fuzz.trimf(rice.universe,[550,800,1200])
rice['vmuch'] = fuzz.trapmf(rice.universe,[800,1200,1500,1500])
pow['vlow'] = fuzz.trimf(pow.universe,[10,10,30])
pow['low'] = fuzz.trimf(pow.universe,[10,30,50])
pow['medium'] = fuzz.trimf(pow.universe,[30,50,70])
pow['high'] = fuzz.trimf(pow.universe,[50,70,100])
pow['vhigh'] = fuzz.trimf(pow.universe,[70,100,100])
pow.view()
rule1 = ctrl.Rule(time['vfast']&rice['vless'],pow['vlow'])
rule2 = ctrl.Rule(time['vfast']&rice['less'],pow['low'])
rule3 = ctrl.Rule(time['vfast']&rice['avg'],pow['medium'])
rule4 = ctrl.Rule(time['vfast']&rice['much'],pow['high'])
rule5 = ctrl.Rule(time['vfast']&rice['vmuch'],pow['vhigh'])
rule6 = ctrl.Rule(time['fast']&rice['vless'],pow['vlow'])
rule7 = ctrl.Rule(time['fast']&rice['less'],pow['low'])
rule8 = ctrl.Rule(time['fast']&rice['avg'],pow['medium'])
rule9 = ctrl.Rule(time['fast']&rice['much'],pow['high'])
rule10 = ctrl.Rule(time['fast']&rice['vmuch'],pow['vhigh'])
rule11 = ctrl.Rule(time['medium']&rice['vless'],pow['low'])
rule12 = ctrl.Rule(time['medium']&rice['less'],pow['low'])
rule13 = ctrl.Rule(time['medium']&rice['avg'],pow['medium'])
rule14 = ctrl.Rule(time['medium']&rice['much'],pow['vhigh'])
rule15 = ctrl.Rule(time['medium']&rice['vmuch'],pow['high'])
rule16 = ctrl.Rule(time['slow']&rice['vless'],pow['vlow'])
rule17 = ctrl.Rule(time['slow']&rice['less'],pow['low'])
rule18 = ctrl.Rule(time['slow']&rice['avg'],pow['high'])
rule19 = ctrl.Rule(time['slow']&rice['much'],pow['high'])
rule20 = ctrl.Rule(time['slow']&rice['vmuch'],pow['vhigh'])
rule21 = ctrl.Rule(time['vslow']&rice['vless'],pow['vlow'])
rule22 = ctrl.Rule(time['vslow']&rice['less'],pow['low'])
rule23 = ctrl.Rule(time['vslow']&rice['avg'],pow['high'])
rule24 = ctrl.Rule(time['vslow']&rice['much'],pow['high'])
rule25 = ctrl.Rule(time['vslow']&rice['vmuch'],pow['vhigh'])
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9,rule10, rule11, rule12,rule13, rule14, rule15, rule16, rule17, rule18, rule19,rule20,rule21, rule22, rule23, rule24, rule25])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['time'] = 78
tipping.input['rice'] = 400
tipping.compute()
print(tipping.output['pow'])
pow.view(sim=tipping)
```
